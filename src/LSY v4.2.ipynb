To increase training speed and model accuracy, the images are first cropped to fufill the image with characters and then resized to a 64 by 64 pixel image.

```
# resize the dataframe
from tqdm.auto import tqdm
def resize(df, size=96, need_progress_bar=True):
    resized = {}
    resize_size=96
    if need_progress_bar:
        for i in tqdm(range(df.shape[0])):
            image=df.loc[df.index[i]].values.reshape(137,236)
            _, thresh = cv2.threshold(image, 30, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
            contours, _ = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)[-2:]

            idx = 0 
            ls_xmin = []
            ls_ymin = []
            ls_xmax = []
            ls_ymax = []
            for cnt in contours:
                idx += 1
                x,y,w,h = cv2.boundingRect(cnt)
                ls_xmin.append(x)
                ls_ymin.append(y)
                ls_xmax.append(x + w)
                ls_ymax.append(y + h)
            xmin = min(ls_xmin)
            ymin = min(ls_ymin)
            xmax = max(ls_xmax)
            ymax = max(ls_ymax)

            roi = image[ymin:ymax,xmin:xmax]
            resized_roi = cv2.resize(roi, (resize_size, resize_size),interpolation=cv2.INTER_AREA)
            resized[df.index[i]] = resized_roi.reshape(-1)
    else:
        for i in range(df.shape[0]):
            image=df.loc[df.index[i]].values.reshape(137,236)
            _, thresh = cv2.threshold(image, 30, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
            contours, _ = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)[-2:]

            idx = 0 
            ls_xmin = []
            ls_ymin = []
            ls_xmax = []
            ls_ymax = []
            for cnt in contours:
                idx += 1
                x,y,w,h = cv2.boundingRect(cnt)
                ls_xmin.append(x)
                ls_ymin.append(y)
                ls_xmax.append(x + w)
                ls_ymax.append(y + h)
            xmin = min(ls_xmin)
            ymin = min(ls_ymin)
            xmax = max(ls_xmax)
            ymax = max(ls_ymax)

            roi = image[ymin:ymax,xmin:xmax]
            resized_roi = cv2.resize(roi, (resize_size, resize_size),interpolation=cv2.INTER_AREA)
            resized[df.index[i]] = resized_roi.reshape(-1)
    resized = pd.DataFrame(resized).T
    return resized
```

```
# Build DataGenerator
class MultiOutputDataGenerator(keras.preprocessing.image.ImageDataGenerator):

    def flow(self,
             x,
             y=None,
             batch_size=32,
             shuffle=True,
             sample_weight=None,
             seed=None,
             save_to_dir=None,
             save_prefix='',
             save_format='png',
             subset=None):

        targets = None
        target_lengths = {}
        ordered_outputs = []
        for output, target in y.items():
            if targets is None:
                targets = target
            else:
                targets = np.concatenate((targets, target), axis=1)
            target_lengths[output] = target.shape[1]
            ordered_outputs.append(output)


        for flowx, flowy in super().flow(x, targets, batch_size=batch_size,
                                         shuffle=shuffle):
            target_dict = {}
            i = 0
            for output in ordered_outputs:
                target_length = target_lengths[output]
                target_dict[output] = flowy[:, i: i + target_length]
                i += target_length

            yield flowx, target_dict
```

```
# Densenet
from tensorflow.keras.layers import MaxPool2D, AveragePooling2D, GlobalAveragePooling2D
def build_dense(IMGSIZE):
    inputs = keras.Input(shape = (IMG_SIZE, IMG_SIZE, 1))
    densenet = Conv2D(filters=3, kernel_size=(2, 2), padding='SAME', activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1))(inputs)

    cur = keras.applications.densenet.DenseNet121(include_top=False, weights='imagenet', input_tensor=None, input_shape=None, pooling=None)(densenet)
    cur = GlobalAveragePooling2D()(cur)
    dense = Dropout(rate=0.4)(cur)
    cur = Dense(1024, kernel_initializer="he_normal", activation='relu')(cur)
    cur = keras.layers.BatchNormalization(momentum=0.15)(cur)
    # cur = Dropout(rate=0.4)(cur)

    # cur = Dense(512, kernel_initializer="he_normal")(cur)
    # cur = keras.layers.BatchNormalization(momentum=0.15)(cur)
    # cur = keras.layers.LeakyReLU(alpha=0.3)(cur)
    dense = Dropout(rate=0.48)(cur)

    # head_root_dense
    # head_root_dense = Dense(512, kernel_initializer="he_normal", activation='relu')(dense)
    # head_root_dense = keras.layers.BatchNormalization(momentum=0.15)(head_root_dense)
    # head_root_dense = Dropout(rate=0.48)(head_root_dense)

    head_root_dense = Dense(512, kernel_initializer="he_normal", activation='relu')(dense)
    head_root_dense = keras.layers.BatchNormalization(momentum=0.15)(head_root_dense)
    head_root_dense = Dropout(rate=0.48)(head_root_dense)

    head_root_dense = Dense(168, activation = 'softmax')(head_root_dense)

    # head vowel dense
    head_vowel_dense = Dense(512, kernel_initializer="he_normal", activation='relu')(dense)
    head_vowel_dense = keras.layers.BatchNormalization(momentum=0.15)(head_vowel_dense)
    # head_vowel_dense = keras.layers.LeakyReLU(alpha=0.3)(head_vowel_dense)
    head_vowel_dense = Dropout(rate=0.45)(head_vowel_dense)


    head_vowel_dense = Dense(11, activation = 'softmax')(head_vowel_dense)

    # head consonant dense
    head_consonant_dense = Dense(512, kernel_initializer="he_normal", activation='relu')(dense)
    head_consonant_dense = keras.layers.BatchNormalization(momentum=0.15)(head_consonant_dense)
    head_consonant_dense = Dropout(rate=0.45)(head_consonant_dense)

    head_consonant_dense = Dense(7, activation = 'softmax')(head_consonant_dense)


    model = keras.models.Model(inputs=inputs, outputs=[head_root_dense, head_vowel_dense, head_consonant_dense])
    return model


model = build_dense(IMG_SIZE)
root_dense = model.layers[-3].name
vowel_dense = model.layers[-2].name
consonant_dense = model.layers[-1].name
print(root_dense, vowel_dense, consonant_dense)
weights = {root_dense: 0.38, vowel_dense: 0.31, consonant_dense:0.31}
model.compile(optimizer=keras.optimizers.Adam(lr=0.001), loss_weights=weights, loss='categorical_crossentropy', metrics=['accuracy'])
```
